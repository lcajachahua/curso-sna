# Primeros pasos con NetworkX

### Crear Grafos Aleatorios (Erdos-Renyi)

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.generators.random_graphs import erdos_renyi_graph
G1 = erdos_renyi_graph(10, 0.5)
print(G1.nodes)
```

```python
print(G1.edges)
```

```python
nx.draw_networkx(G1)
plt.show()
```

```python
G2 = erdos_renyi_graph(10, 0.2)
nx.draw_networkx(G2)
plt.show()
```

```python
G3 = erdos_renyi_graph(10, 0.8)
nx.draw_networkx(G3)
plt.show()
```

### Crear Grafos Watts-Strogatz

```python
ws = nx.watts_strogatz_graph(20, 5, 0.1)
nx.draw(ws)
```

```python
ws = nx.watts_strogatz_graph(100, 10, 0.1)
nx.draw(ws)
```

### Crear Grafos Barabasi - Albert

```python
ba = nx.barabasi_albert_graph(20, 2)
nx.draw(ba)
```

```python
ba = nx.barabasi_albert_graph(100, 5)
nx.draw(ba)
```

### Otros Grafos particulares

### Matrices Importantes

Matriz de Adyacencia

```python
import scipy as sp
A = nx.adjacency_matrix(G2)
print(A.todense())
```

```python
import scipy as sp
A = nx.adjacency_matrix(G3)
print(A.todense())
```

Matriz de Incidencia

```python
I = nx.incidence_matrix(G2)
print(I.todense())
```

```python
I = nx.incidence_matrix(G3)
print(I.todense())
```

Matriz de Atributos

```python
T = nx.attr_matrix(G3)
print(T)
```
