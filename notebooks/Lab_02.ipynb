# Primeros pasos con NetworkX

### Crear un Grafo Aleatorio (Erdos-Renyi)

```python
from networkx.generators.random_graphs import erdos_renyi_graph
n = 10
p = 0.5
G1 = erdos_renyi_graph(n, p)
print(G1.nodes)
```

```python
print(G1.edges)
```

```python
pos = nx.spring_layout(G1)
nx.draw_networkx(G1, pos)
plt.title("Grafo Erdos Renyi p=0.5")
plt.gcf().canvas.set_window_title("")
plt.show()
```

```python
n = 10
p = 0.2
G2 = erdos_renyi_graph(n, p)
pos = nx.spring_layout(G2)
nx.draw_networkx(G2, pos)
plt.title("Grafo Erdos Renyi p=0.2")
plt.gcf().canvas.set_window_title("")
plt.show()
```
